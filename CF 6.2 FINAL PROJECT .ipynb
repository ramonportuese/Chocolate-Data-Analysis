```python
# flin 23
# So, when you run this line of code, the computer will go through your
# list of chocolates, pick out the ones that have a cocoa percentage 
# between 60% and 75%, and make a new list just for those chocolates. 
# It's like creating a special list of chocolates with a specific cocoa level.

# *** Question: Terry is this wrong? 
choc_bar_ratings_df_clean.loc[(choc_bar_ratings_df_clean['Cocoa Percent'] >=0.6) & (choc_bar_ratings_df_clean['Cocoa Percent'] <0.75), 'Cocoa Level']
```

```python

# # So, What this code is doing is having the computer go through # flin 24 
# the list of chocolates, pick out the ones that have a cocoa percentage
# between 75% and less than 100%, and make a new list just for those chocolates. 
# It's like creating a special list of chocolates with a higher cocoa level.

choc_bar_ratings_df_clean.loc[(choc_bar_ratings_df_clean['Cocoa Percent'] >=0.75) & (choc_bar_ratings_df_clean['Cocoa Percent'] <1), 'Cocoa Level']
```
