```python
# now I am going to seperate the Ingredients column into two seperate columns 
# the reason behind this is because the Ingredients column has integers and strings 
# and I want more difintive clarity in my data set. So, the Ingredients column will 
# be split into a new "Num of Ingredients" column which is represented by an Integer/  #
# and another column that is "Specific Ingredients"
```

```python
# Now I am going to be Sepperating the 'Ingredients' column into two separate columns'Num_of_Ingredients' and 'Specific Ingredients'
choc_bar_ratings_df[['Num_of_Ingredients', 'Specific Ingredients']] = choc_bar_ratings_df['Ingredients'].str.split('-', expand=True)

# Remove leading and trailing whitespace from both columns
choc_bar_ratings_df['Num_of_Ingredients'] = choc_bar_ratings_df['Num_of_Ingredients'].str.strip()
choc_bar_ratings_df['Specific Ingredients'] = choc_bar_ratings_df['Specific Ingredients'].str.strip()
```

```python
# ***Note: Flin 30 might need to go back and delete ? is this correct? 
# creating dictionary and map to replace NaN specific to each country ('Compnay Location')
replacement_values = {
    'Argentina': 4,  # Replace NaN for Argentina with 4, etc.
    'Australia': 3,  
    'Belgium': 4,
    'Brazil': 3,  
    'Canada': 3,
    'Colombia': 4,
    'Ecuador': 4,
    'France': 4,
    'Germany': 3,
    'Italy': 3,
    'Japan': 3,
    'Madagascar': 4,
    'New Zealand': 2,
    'Peru': 3,
    'Poland': 2,
    'Russia': 2,
    'Sao Tome & Principe': 3,
    'South Korea': 2,
    'Spain': 3,
    'U.K.': 3,
    'U.S.A.': 3
}
```

```python
# ***Note: Flin 31 might need to go back and delete ? is this correct? 
# Define a custom function to replace NaN values
def replace_nan(row):
    location = row['Company Location']
    replacement = replacement_values.get(location)
    if pd.isna(row['Num_of_Ingredients']):
        return replacement
    else:
        return row['Num_of_Ingredients']

# Apply the custom function to replace NaN values in 'Num_of_Ingredients'
choc_bar_ratings_df['Num_of_Ingredients'] = choc_bar_ratings_df.apply(replace_nan, axis=1)
```
